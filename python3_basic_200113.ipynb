```
#딕셔너리 연습문제
apps_of_ages = {"4+":4493, "9+":987, "12+":1155, "17+":622}
total_apps = 7197
ratios = {}
percentages = {}
for age in apps_of_ages:
  ratios[age] = apps_of_ages[age] / total_apps
  percentages[age] = ratios[age] * 100
print(ratios)
print(percentages)
```

```
def describe_pet(animal_type, pet_name):
  print("\nI have a " + animal_type+".")
  print("My " + animal_type + "'s name is " + pet_name.title() + ".")

describe_pet("hamster", "harry")
describe_pet("dog", "willie")

#키워드를 이용한 파라미터 전달
describe_pet(animal_type="hamster", pet_name="harry")
describe_pet(pet_name="willie", animal_type="dog")
```

```
def get_formatted_name(first_name, last_name, middle_name=''):
  if middle_name:
    full_name = first_name + ' ' + middle_name + ' ' + last_name
  else:
    full_name = first_name + ' ' + last_name
  return full_name.title()

musician = get_formatted_name('jimi', 'handrix')
print(musician)

musician = get_formatted_name('louis', 'armstrong', 'daniel')
print(musician)
```

```
#딕셔너리 리턴
def build_person(first_name, last_name):
  person = {'fist' : first_name, 'last' : last_name}
  return person

musician = build_person('jimi', 'handrix')
print(musician)
```

```
#함수 연습문제
def square(x):
  return x ** 2
print(square(2)-4)
print(square(10))
print(square(16))
print("\n")

def avg1(a, b, c):
  return (a+b+c) / 3
print(avg1(90,70,80))
print("\n")

def avg2(*scores):
  sum = 0
  for score in scores:
    sum+=score
  return sum / len(scores)
print(avg2(84,69,75))
print(avg2(87,90,89,95))
print("\n")

def grade(*scores):
  sum = 0
  for score in scores:
    sum+=score
  average = sum / len(scores)
  if average >= 90:
    return 'A'
  elif average >= 80:
    return 'B'
  elif average >= 70:
    return 'C'
  else:
    return 'F'

print(grade(84,69,75))
print(grade(87,90,89,95))
```

```
from bs4 import BeautifulSoup
import requests

soup = BeautifulSoup(requests.get("http://www.consumerreports.org/cro/a-to-z-index/products/index.htm").text, 'html.parser')

attr = {"class": "crux-body-copy"}
list_div = soup.find_all("div", attrs=attr)
for div in list_div:
  ele_a = div.find('a')
  if 'href' in ele_a.attrs:
    print("url : http://www.consumerreports.org" + str(ele_a.attrs['href']))
    print("goods_name : " + str(ele_a).split("\n")[1])
  print("\n")
```

```
from bs4 import BeautifulSoup
import requests

soup = BeautifulSoup(requests.get("https://codingbat.com/java/Warmup-1").text, 'html.parser').find("div", attrs={"class":"indent"})

for new_url in soup.find_all("td"):
  ele_a = new_url.find('a')
  if 'href' in ele_a.attrs:
    new_soup = BeautifulSoup(requests.get("https://codingbat.com/"+str(ele_a.attrs['href'])).text, 'html.parser').find("table", attrs={"border":"0"})
    print(new_soup.find("p",attrs={"class":"max2"}).contents[0])
    print()
    for br in new_soup.find_all("br"):
      print(br.next)
    print("\n")
```
